#Ranking

```python
import torch
from transformers import DistilBertTokenizer, DistilBertModel

# Load pre-trained DistilBERT model and tokenizer
model_name = "distilbert-base-uncased"
tokenizer = DistilBertTokenizer.from_pretrained(model_name)
model = DistilBertModel.from_pretrained(model_name)

# Sample job description and CVs
job_description = "Looking for a software engineer with experience in Python with experience in Keras and tensorflow and web development Web technolgies must include HTML,CSS,React,Angular."
cvs = [
    "Experienced software engineer skilled in Python and web development.",
    "Data scientist with expertise in machine learning and Python programming with Keras and Tensorflow.",
    "Front-end developer with experience in HTML,CSS,JavaScript and React.I also have knowledge in Python with Keras and tensorflow.",
]

def rank_cvs(job_description, cvs):
    # Tokenize the job description and CVs
    inputs = tokenizer([job_description] + cvs, return_tensors="pt", padding=True, truncation=True)
    with torch.no_grad():
        # Get embeddings for the job description and CVs
        outputs = model(**inputs)
        embeddings = outputs.last_hidden_state[:, 0, :]

    # Calculate cosine similarity between the job description and CVs
    similarity_scores = torch.nn.functional.cosine_similarity(embeddings[0].unsqueeze(0), embeddings[1:])

    # Sort CVs based on similarity score in descending order
    ranked_cvs = [cv for _, cv in sorted(zip(similarity_scores, cvs), reverse=True)]

    return ranked_cvs

# Rank CVs based on job alignment
ranked_cvs = rank_cvs(job_description, cvs)
for i, cv in enumerate(ranked_cvs):
    print(f"Rank {i+1}: {cv}")
```

#Screening questions

```python
class ScreeningQuestion:
    def __init__(self, question, weight):
        self.question = question
        self.weight = weight

def ask_screening_questions(candidate_name):
    questions = [
        ScreeningQuestion("Do you have experience in Python?", weight=5),
        ScreeningQuestion("Have you worked on web development projects?", weight=4),
        ScreeningQuestion("Do you possess strong communication skills?", weight=3),
        ScreeningQuestion("Have you ever managed a team?", weight=4),
        ScreeningQuestion("Are you familiar with cloud technologies?", weight=2),
    ]

    total_weight = sum(question.weight for question in questions)

    print(f"Screening questions for {candidate_name}:")
    for i, question in enumerate(questions, start=1):
        print(f"{i}. {question.question}")

    score = 0
    for question in questions:
        answer = input(f"Enter your answer to Question {questions.index(question) + 1}: ").strip().lower()
        if answer == "yes":
            score += question.weight

    normalized_score = score / total_weight
    print(f"\n{candidate_name}'s normalized screening score: {normalized_score:.2f}")

    # Add further evaluation logic based on the screening score
    if normalized_score >= 0.7:
        print(f"{candidate_name} qualifies for the next round.")
    else:
        print(f"{candidate_name} did not meet the required screening threshold.")

# Sample candidates and their screening questions
candidates = [
    ("Bhuvana Gayathri",),
    ("Sravani M",),
]

# Ask screening questions for each candidate
for candidate in candidates:
    ask_screening_questions(*candidate)
```

#First round interview

```python
class Candidate:
    def __init__(self, name):
        self.name = name
        self.screening_score = 0
        self.responses = {}

    def answer_question(self, question, answer):
        self.responses[question] = answer

    def calculate_screening_score(self, total_weight):
        total_score = sum(self.responses.get(q, 0) * q.weight for q, answer in self.responses.items())
        self.screening_score = total_score / total_weight

def conduct_first_round_interview(candidate, screening_questions):
    print(f"Interviewing {candidate.name} for the first round.")

    total_weight = sum(question.weight for question in screening_questions)

    for question in screening_questions:
        answer = input(f"{question.question} ")
        answer = 1 if answer.lower() == "yes" else 0
        candidate.answer_question(question, answer)

    candidate.calculate_screening_score(total_weight)
    print(f"\n{candidate.name}'s normalized screening score: {candidate.screening_score:.2f}")

# Sample candidates and their screening questions
candidates = [
    Candidate("Bhuvana Gayathri"),
    Candidate("Sravani M"),
]

screening_questions = [
    ScreeningQuestion("Do you have experience in Python?", weight=5),
    ScreeningQuestion("Have you worked on web development projects?", weight=4),
    ScreeningQuestion("Do you possess strong communication skills?", weight=3),
    ScreeningQuestion("Have you ever managed a team?", weight=4),
    ScreeningQuestion("Are you familiar with cloud technologies?", weight=2),
]

# Conduct the first-round interview for each candidate
for candidate in candidates:
    conduct_first_round_interview(candidate, screening_questions)

# Evaluate candidates for the next in-person round
qualified_candidates = [c for c in candidates if c.screening_score >= 0.7]

print("\nQualified candidates for the next round:")
if qualified_candidates:
    for candidate in qualified_candidates:
        print(f"- {candidate.name}")
else:
    print("No candidates qualified for the next round.")
```
